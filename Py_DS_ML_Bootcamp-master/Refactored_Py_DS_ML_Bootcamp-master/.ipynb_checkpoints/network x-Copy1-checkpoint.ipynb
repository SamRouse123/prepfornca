```python
import networkx as nx
import matplotlib.pyplot as plt

#follwing gude from youtube video
```

By definition, a Graph is a collection of nodes (vertices) along with identified pairs of nodes (called edges, links, etc)

```python
G = nx.Graph() # Simple undirected graph 

G2 = nx.DiGraph() # directed graph 

G3 = nx.MultiGraph() # graph where we have multiple edges between nodes

G4 = nx.MultiDiGraph() # directed graph with multiple edges
```

```python
G.add_edge(1,2)
nx.draw_spring(G, with_labels=True)
```

```python
G.add_edge(2, 3, weight=0.9)
nx.draw_spring(G, with_labels=True)
```

```python
G.add_edge("A", "B")
nx.draw_spring(G, with_labels=True)
plt.show()
```

```python
edge_list = [(1,2),(2,3),(3,5),(4,6),(6,7)]
G = nx.Graph()
G.add_edges_from(edge_list)
nx.draw_spring(G, with_labels=True)
plt.show()
```
